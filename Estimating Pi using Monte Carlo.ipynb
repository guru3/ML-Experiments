### How do we do this ? Say we have circle of diameter 1 centered at origin, with a square of side 1 and centered at origin. If we pick a random point in square, we have probability equalling Area(circle)/Area(square) that the point lies inside the circle! Now if we do this for tens of thousands of points, and calculate probability as (total point inside circle)/(total points) and compare that with what we got in terms of areas, we can get : Pi = (4*total points inside circle)/total points.
### Let's try it out

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def simulate(totalPoints : int):
    randomPoints = np.random.rand(totalPoints, 2) #2-dimension
    insideCircle = randomPoints[ randomPoints[:,0]**2 + randomPoints[:,1]**2 < 1 ]
    estimate = 4*len(insideCircle)/totalPoints
    absolute_difference = abs( math.pi - estimate )
    plt.figure(figsize=(4,4))
    plt.scatter(randomPoints[:,0], randomPoints[:,1], c='red', s=0.5)
    plt.scatter(insideCircle[:,0], insideCircle[:,1], c='blue', s=0.5)
    plt.show()
    print('Estimate of pi using %d points is %5.8f, with difference of %2.8f' % (totalPoints, estimate, absolute_difference))
```

```python
simulate(200)
```

```python
simulate(2000)
```

```python
simulate(10000)
```

```python
simulate(50000)
```

```python
simulate(100000)
```

```python
simulate(1000000)
```

```python
simulate(5000000)
```

```python
runSimulationTill = 500000;
plt.figure(figsize=(10,5))
plt.axhline(y=math.pi, color='r', linestyle='-', label='value of pi')
randomPoints = np.random.rand(runSimulationTill, 2) #2-dimension
X = [];
Y = [];

totalInside = 0;
for totalPoints in range(runSimulationTill):
    isInside = randomPoints[totalPoints,0]**2 + randomPoints[totalPoints,1]**2 < 1;
    totalInside += isInside
    estimate = 4*totalInside/(totalPoints+1)
    if totalPoints > 500: #once we get decent estimate
        X.append(totalPoints)
        Y.append(estimate)
        
plt.plot(X, Y, label='estimate');
plt.legend()
plt.xlabel('Total simulations')
plt.show()
```

### We see that as we increase number of simulations we go closer to the actual value of pi..until we reach too close to it that few thousand of iterations don't really get us any closer.
### Overall we have tried to show how we can use monte carlo method to get some meaningful information purely based on use randomly generated data and some logic around consuming that randomness.
